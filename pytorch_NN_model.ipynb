What is torch.nn?

torch.nn is a PyTorch module that helps us build neural networks easily.

Before, you made everything manually:

self.weight = torch.rand(...)
self.bias = torch.zeros(...)

z = x * self.weight + self.bias

y_pred = torch.sigmoid(z)

loss = ...

But torch.nn gives ready-made tools for these things.

In simple words:

torch.nn is like a toolbox for building neural networks.

**1. nn.Module**

nn.Module is the main base class for making a model.

Logic example:

A neural network model is like a machine.

Input data → Model/Machine → Output prediction

In PyTorch, we create that machine using:

import torch

import torch.nn as nn

class MyModel(nn.Module):

    def __init__(self):
        super().__init__()

    def forward(self, x):
        return x

nn.Module helps PyTorch understand:

These are my layers.

These are my weights.

This is my forward pass.

**2. Layers**

Layers are parts of the model.

Example:

nn.Linear(30, 1)

This means:

Input features = 30

Output = 1

This one line replaces your manual code:

self.weight = torch.rand(30, 1)

self.bias = torch.zeros(1)

z = x @ self.weight + self.bias

So instead of manually creating weight and bias, we can write:

self.linear = nn.Linear(30, 1)

Logical example:

Suppose you have patient data:

age

blood pressure

cholesterol

sugar level

...

If there are 30 features, then:

nn.Linear(30, 1)

means:

Take 30 inputs and produce 1 output prediction.

**3. Activation Functions**

Activation functions help the model learn complex patterns.

Example:

nn.Sigmoid()

Sigmoid converts output into a value between 0 and 1.

0.8 = 80% chance class 1

0.2 = 20% chance class 1

In your manual model, you wrote:

y_pred = torch.sigmoid(z)

Using nn, we can write:

self.sigmoid = nn.Sigmoid()

Other common activations:

nn.ReLU()

nn.Tanh()

nn.Sigmoid()

Simple logic:

Without activation, model is too simple.
With activation, model can learn more difficult patterns.

**4. Loss Functions**

Loss function tells the model:

How wrong is your prediction?

In your manual code, you wrote binary cross entropy loss manually:

loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()

But PyTorch already gives ready-made loss functions.

Example:

loss_fn = nn.BCELoss()

or better:

loss_fn = nn.BCEWithLogitsLoss()

Logical example:

Actual answer:

y = 1

Model prediction:

y_pred = 0.8

Loss will be small because prediction is close.

But if model predicts:

y_pred = 0.1

Loss will be high because prediction is wrong.

**5. nn.Sequential**

nn.Sequential means:

Put layers one after another in order.

Example:

model = nn.Sequential(
    nn.Linear(30, 10),

    nn.ReLU(),

    nn.Linear(10, 1),

    nn.Sigmoid()
)

This means:

Step 1: Take 30 inputs and convert to 10 values
Step 2: Apply ReLU activation
Step 3: Convert 10 values to 1 output
Step 4: Apply Sigmoid to get probability

Simple logic:

It is like a pipeline:

Input → Layer 1 → Activation → Layer 2 → Output

**6. Dropout and BatchNorm**

These are used to improve training.

Dropout

nn.Dropout(0.5)

Dropout randomly turns off some neurons during training.

Logic example:

Imagine students always depend on one smart student in a group.

Dropout says:

Sometimes remove that smart student, so everyone learns.

In neural networks, dropout helps prevent overfitting.

Overfitting means:

Model memorizes training data but performs badly on new data.

BatchNorm

nn.BatchNorm1d()

Batch normalization helps keep values stable during training.

Simple logic:

If data values are very different:

age = 25

salary = 50000

marks = 80

The model may struggle.

BatchNorm helps normalize/adjust values so training becomes smoother.

Simple Comparison with Your Manual Model

Your manual model:

self.weight = torch.rand(...)

self.bias = torch.zeros(...)

z = x @ self.weight + self.bias

y_pred = torch.sigmoid(z)

Using torch.nn:

self.linear = nn.Linear(30, 1)

self.sigmoid = nn.Sigmoid()

Much cleaner.

Simple Full Example

import torch

import torch.nn as nn

class MySimpleNN(nn.Module):

    def __init__(self):

        super().__init__()

        self.linear = nn.Linear(30, 1)

        self.sigmoid = nn.Sigmoid()

    def forward(self, x):

        z = self.linear(x)

        y_pred = self.sigmoid(z)

        return y_pred

Use it like this:

model = MySimpleNN()

y_pred = model(x_train_tensor)

Notice:

model(x_train_tensor)

automatically calls:

forward()

Easy Memory Trick

nn.Module     = base class for model

nn.Linear     = layer with weight and bias

nn.Sigmoid    = converts output into probability

nn.ReLU       = helps model learn complex patterns

nn.BCELoss    = calculates prediction error

nn.Sequential = stacks layers in order

nn.Dropout    = prevents overfitting

```python
import torch # -> this is library
import torch.nn as nn # we are loading the nn
```

```python
class Model(nn.Module): # we are creating the class in which loading the nn module

  def __init__(self , num_features): # its a constructor and is the parametor is num_feature mean number of features like columns

    super().__init__()  # its a inheritence childe from the parents super().__init__() ka matlab hai:

#Parent class nn.Module ko activate karo taake PyTorch model ke layers aur parameters ko track kar sake.

    self.linear = nn.Linear(num_features , 1) # define the number of feature(features) and output ( 1 )

    self.sigmoid = nn.Sigmoid() # getting the predictions with sigmoit function

  def forward(self , features): # create a forword function

    out = self.linear(features) # here we are doing forward pass like z = (x * w) + b

    out = self.sigmoid(out) # pred = sigmoid(z )

    return out
```

```python
 # create a dataset
features = torch.rand(10 , 5)


 # create model
model = Model(features.shape[1])

 # call model for forword pass
#model.forword(features) -> hum asy b ker skty hn lakin

model(features) # this way is better then that way becouse this function call as a construtor automatic call just creat the object and pass the feature
```

The rand function is used to generate random values. When we execute the same code multiple times, the output becomes different because the random function generates a new value each time.


```python
#useing manual seed
# useing seed
torch.manual_seed(100) #-> 100 bar ap excecute kro gay tu same he aoutput nikly gi is k bad value change ho jay kry gi
torch.rand(3,3)
```

magine you are working in a team and using some code. After some time, your team member may use the same code and want to get the same generated values as you. In this case, we use a seed function.


```python
#useing tensor
torch.tensor([[1,2,3],[4,5,6]])
```

```python
# other ways
#arange

print(f"useing arange -> ",torch.arange(0,10))

#arange with jump -> 2

print(f"useing arange -> ",torch.arange(0,10 , 2))

# linespace
print(f"useing linespave -> ",torch.linspace(4,5,10))

# eye -> identity matrix

print(f"useing eye #",torch.eye(3,4))


# full
print(f"full useing -> ",torch.full((3,4),5))
```
